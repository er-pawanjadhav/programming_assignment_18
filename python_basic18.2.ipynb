```python
a ='HeLlo World '
a = a.split(" ")
d = []
s = ''

def rem_s(a):
    """this function remove blank
    space between the word
    """
    for i in a:
        if i!='':
            d.append(i)
    return d
d = rem_s(a)
        
def rev(a):
    """this function 
    reverse world"""
    s=''
    for j in a:
        s=j+s
    return s
for i in d:
    e = rev(i)
    s= e +' '+s
def lower_upper(s): 
    """this function make lower letter 
    to upper and viceversa"""
    c = ''
    for i in s:
        if i.islower():
            i = i.upper()
        else:
            i = i.lower()
        c = c+i
    return c
print(lower_upper(s))
```
